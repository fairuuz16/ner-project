### Load Dataset

```python
df = pd.read_csv("./dataset.csv", sep=',')
df.head()
```

#### Tokenisasi

```python
df['judul'] = df['judul'].fillna('')
df['isi_berita'] = df['isi_berita'].fillna('')

# Tokenisasi 
df['judul_tokens'] = df['judul'].apply(word_tokenize)
df['isi_tokens'] = df['isi_berita'].apply(word_tokenize)

# Hasil tokenisasi
df[['judul', 'judul_tokens', 'isi_berita', 'isi_tokens']].head()
```

#### Casefolding


```python
# Casefolding
df['judul_tokens_lower'] = df['judul_tokens'].apply(lambda tokens: [token.lower() for token in tokens])
df['isi_tokens_lower'] = df['isi_tokens'].apply(lambda tokens: [token.lower() for token in tokens])

# Hasilnya
df[['judul', 'judul_tokens_lower', 'isi_tokens_lower']].head()
```

#### Stemming

```python
from Sastrawi.Stemmer.StemmerFactory import StemmerFactory

# Stemmer Indonesia
factory = StemmerFactory()
stemmer = factory.create_stemmer()

# Stemming judul dan isi berita
df['judul_stemmed'] = df['judul_tokens_lower'].apply(lambda tokens: [stemmer.stem(token) for token in tokens])
df['isi_stemmed'] = df['isi_tokens_lower'].apply(lambda tokens: [stemmer.stem(token) for token in tokens])

# Hasil
df[['judul', 'judul_stemmed', 'isi_stemmed']].head()
```
